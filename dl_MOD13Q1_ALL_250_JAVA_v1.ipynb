```
# Import the Earth Engine Python Package
import ee
import time
import os
import math
from datetime import datetime


# Initialize the Earth Engine object, using the authentication credentials.

ee.Initialize()


def getMOD13Q1(img):
	NDVI = img.select(['NDVI']).toInt16()
	EVI = img.select(['EVI']).toInt16()
	DayOfYear = img.select(['DayOfYear']).toInt16()
	SummaryQA = img.select(['SummaryQA']).toInt16()
	return img.select([]).addBands([NDVI,EVI,DayOfYear,SummaryQA]).select([0,1,2,3],['NDVI','EVI','DayOfYear','SummaryQA'])  


java_region =[[[104.87671324999405, -6.617130502942074],[105.26123473436905, -7.216977733951505],[107.00806090624405, -7.772491468163925],[109.24927184374405, -8.055415211422698],[112.06177184374405, -8.58807517471107],[114.70947692186905, -9.06574699261073],[114.72046324999405, -7.696286806092489],[114.45679137499405, -6.715338959664886],
[109.91943785936905, -5.950995062539529],[106.53564879686905, -5.601218082333037],[105.04150817186905, -6.180416120421386]]]
		  
		  
java_poly = ee.Geometry.Polygon(java_region)

MOD13Q1_Collection = ee.ImageCollection('MODIS/006/MOD13Q1').filterDate('2018-01-01','2020-10-01')
MOD13Q1_Collection = MOD13Q1_Collection.filterBounds(java_poly)
MOD13Q1_Collection = MOD13Q1_Collection.sort('DATE_ACQUIRED')
MOD13Q1_Collection = MOD13Q1_Collection.select(['NDVI','EVI','DayOfYear','SummaryQA'])
MOD13Q1_Collection = MOD13Q1_Collection.map(getMOD13Q1)

colList = MOD13Q1_Collection.toList(4000)
n = colList.size().getInfo();
print (n)
for j in range(0, n):
  img = ee.Image(colList.get(j))
  id = img.id().getInfo()
  filename = 'MOD13Q1-JAVA-250m-utm-' + id
  task_config = {
    'scale': 250,
		'crs' : 'EPSG:32749',
		'region': java_region,
	  'driveFolder': 'MOD13Q1-JAVA-250m-utm'
	}
  task = ee.batch.Export.image(img, filename, task_config)
  task.start()
  print (datetime.now().strftime("%Y-%m-%d %H:%M:%S") + ' ->  '  + str(j+1) + ' - ' + filename)

```
